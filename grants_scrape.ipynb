```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL and pages to scrape
base_url = "https://nij.ojp.gov/funding/awards/list"
pages_to_scrape = 10  # first 10 pages

# List to store all award data
all_awards = []

# Loop through pages
for page in range(pages_to_scrape):
    if page == 0:
        url = base_url
    else:
        url = f"{base_url}?page={page}#awards-awards-list-block-5jx36lxphbxtj8vi"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the table rows
    rows = soup.select("tr")  # all rows in the table
    
    for row in rows[1:]:  # skip header row
        cells = row.find_all("td")
        if len(cells) == 8:
            award = {
                "Fiscal Year": cells[0].get_text(strip=True),
                "Title": cells[1].get_text(strip=True),
                "Funding Opportunity": cells[2].get_text(strip=True),
                "Awardee": cells[3].get_text(strip=True),
                "Awardee State": cells[4].get_text(strip=True),
                "Award Number": cells[5].get_text(strip=True),
                "Award Amount": cells[6].get_text(strip=True),
                "Award Status": cells[7].get_text(strip=True),
            }
            all_awards.append(award)

print(f"Total awards scraped: {len(all_awards)}")
```

```python
# Convert to a DataFrame
df = pd.DataFrame(all_awards)

# Preview the first 5 rows
df.head()
```
