```python
import requests
from bs4 import BeautifulSoup

url = "https://doc.la.gov/about-the-dpsc/facility-locations/"
response = requests.get(url)
print(response.text[:500])  # print the first 500 characters
```

```python
soup = BeautifulSoup(response.text, "html.parser")

# Find all facility blocks
facility_blocks = soup.find_all("a", class_="pt-2 pb-3 px-4 d-block")

# Extract facility names
facility_names = [block.find("h3").text.strip() for block in facility_blocks]

# Preview the first 10 names
facility_names[:10]
```

```python
import csv

# Save facility names to CSV
with open("facilities.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Facility Name"])  # header
    for name in facility_names:
        writer.writerow([name])

print("Saved to facilities.csv")
```
